# FIFA 2019 Data Analysis

```python
# Importing required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
pd.set_option('display.max_columns',87)
```

#### 1.Getting the dataset

```python
fifa_df = pd.read_csv('data.csv')
```

```python
fifa_df.head()
```

#### 2.1 Determining the number of attributes in the dataset

```python
fifa_df.columns
```

#### 2.2 Cleaning the dataset.

```python
missing_data = pd.isna(fifa_df.columns).sum()
missing_data
```

### 3.Exploratory Analysis and Visualization

```python
plt.figure(figsize=(25,25))
sns.heatmap(fifa_df.corr(),annot=True,fmt='.1f')
plt.title("Correlation between the properties of attributes of football players")
plt.show()
```

#### 3.2 Nation wise list of players contributing to football

```python
counts_Nationality = fifa_df['Nationality'].value_counts()
counts_Nationality = counts_Nationality.reset_index()
counts_Nationality.columns = ["Nationality","Counts"]
#counts_Nationality
```

```python
plt.figure(figsize=(10,30))
sns.barplot(x='Counts',y='Nationality',data=counts_Nationality)
plt.title("Nation wise Players Counts")
plt.show()
```

### 3.3 Comparison of positions to number of players

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Position',data=fifa_df)
plt.xlabel("Positions")
plt.ylabel("No of Players")
plt.title("Representation of Players in different Positions")
plt.show()
```

### 3.4 Preferred foot of players

```python
preferred_foot =  fifa_df['Preferred Foot'].value_counts().reset_index()
preferred_foot.columns = ['Foot','Total Players']
preferred_foot = preferred_foot.set_index('Foot')
preferred_foot
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x=fifa_df['Preferred Foot'])
plt.title("Preferred Foot")
plt.show()
```

### 3.5 Player's Overall Rating Distribution

```python
plt.hist(fifa_df['Overall'])
plt.title("Players with Overall Rating")
plt.xlabel("Overall Ratings")
plt.ylabel("No of Players")
plt.show()
```

5.Which club has the highest Overall Rating?

```python
highest_overall_club = fifa_df.groupby('Club')['Overall'].mean()
highest_overall_club = highest_overall_club.reset_index()
highest_overall_club = highest_overall_club.sort_values('Overall',ascending=False)
highest_overall_club.head(10)
```

```python
top_ten_club = highest_overall_club.head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_ten_club['Overall'],
            y=top_ten_club['Club'],
           palette='inferno')
plt.title("Top 10 Best Overall Clubs")
plt.show()
```
